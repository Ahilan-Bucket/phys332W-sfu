## Q1 (10/10)

1. 2 measurements $x_1$ & $x_2$, each with random uncertainty, common systematic uncertainty that introduces correlation between them.

**Suppose:** $x_1 = x_1^R + a$, $x_2 = x_2^R + a$ in terms of underlying variables $x_1^R$ & $x_2^R$ with uncertainty $\sigma_1$ & $\sigma_2$ and common variable $a$ with uncertainty $\sigma_a$.  
$\rightarrow$ Assume $x_1^R$, $x_2^R$ & $a$ all have zero mean value and all have normally distributed random error independent of one another.

**Find:** An expression for the variance of $x_1$ & $x_2$, their covariance & the correlation coefficient $\rho_{12}$ in terms of $\sigma_1$, $\sigma_2$ & $\sigma_a$.

$\Rightarrow$ The variance of $x_1$ & $x_2$:  
$\text{Var}(x_1) = \langle (x_1 - \langle x_1 \rangle)^2 \rangle$  
$\text{Var}(x_2) = \langle (x_2 - \langle x_2 \rangle)^2 \rangle$

**Note:** Problem states "$x_1^R, x_2^R, a$" have zero mean value, i.e. take mean, you get $\langle x_1^R \rangle = 0 = \langle x_2^R \rangle = \langle a \rangle$.

① Variable $x_1^R$ & $x_2^R$ have uncertainty $\sigma_1$ & $\sigma_2$. Common variable $a$ of uncertainty $\sigma_a$.  
i.e., $\langle(x_1^R)^2\rangle = \sigma_1^2$, $\langle(x_2^R)^2\rangle = \sigma_2^2$, $\langle a^2\rangle = \sigma_a^2$  
$\text{Cov}(x_1^R, x_1^R) = \sigma_1^2$, $\text{Cov}(x_2^R, x_2^R) = \sigma_2^2$, $\text{Cov}(a,a) = \sigma_a^2$

② $x_1^R, x_2^R$ & $a$ are independent random variables with zero mean,  
i.e., $\langle x_1^R a\rangle = \langle x_2^R a\rangle = \langle x_1^R x_2^R\rangle = 0$

$\therefore \text{Var}(x_1) = \langle(x_1 - \langle x_1 \rangle)^2\rangle = \langle(x_1^R + a - \langle x_1^R + a \rangle)^2\rangle$

$\rightarrow$ Here $\langle x_1 \rangle = \langle x_1^R + a \rangle = \langle x_1^R \rangle + \langle a \rangle = 0$ from ①

So $\text{Var}(x_1) = \langle(x_1^R + a)^2\rangle$  
$= \langle(x_1^R)^2 + 2x_1^R a + a^2\rangle$  
$= \langle(x_1^R)^2\rangle + 2\langle x_1^R a\rangle + \langle a^2\rangle$  
$= (\sigma_1^R)^2 + 0 + \sigma_a^2$  
(from ① and ②)

$\therefore \mathbf{\text{Var}(x_1) = (\sigma_1^R)^2 + \sigma_a^2}$  
Likewise, $\mathbf{\text{Var}(x_2) = \langle(x_2^R + a)^2\rangle = (\sigma_2^R)^2 + \sigma_a^2}$

Next, $\text{Cov}(x_1, x_2) = \langle(x_1 - \langle x_1 \rangle)(x_2 - \langle x_2 \rangle)\rangle$  
$= \langle x_1, x_2 \rangle = \langle(x_1^R + a)(x_2^R + a)\rangle$  
$= \langle x_1^R x_2^R + x_1^R a + x_2^R a + a^2\rangle$  
$= \langle x_1^R x_2^R\rangle + \langle x_1^R a\rangle + \langle x_2^R a\rangle + \langle a^2\rangle$  
$= 0 + 0 + 0 + \sigma_a^2$  
(from ① and ②)

$\therefore \mathbf{\text{Cov}(x_1, x_2) = \sigma_a^2}$

**Correlation coefficient:**  
$$ \rho_{12} = \frac{\text{Cov}(x_1, x_2)}{\sqrt{\text{Var}(x_1)\,\text{Var}(x_2)}} = \frac{\sigma_a^2}{\sqrt{[(\sigma_1^R)^2 + \sigma_a^2][(\sigma_2^R)^2 + \sigma_a^2]}} $$

$$ \boxed{\rho_{12} = \frac{\sigma_a^2}{[(\sigma_1^R)^2(\sigma_2^R)^2 + (\sigma_1^R\sigma_a)^2 + (\sigma_2^R\sigma_a)^2 + \sigma_a^4]^{1/2}}} $$

**✓ Verification:** $\sigma_a$ is the common thread, a shared systematic error. If $\sigma_a \rightarrow 0$, then correlation coefficient (the off-diagonal quantity that tells how related quantities are) must go to zero. ✓ Yes it does.  
If $\sigma_a \rightarrow$ large: $\rho_{12} \approx \sigma_a^2/(\sigma_a^2) = 1$ (perfect correlation!) Yes matches intuition.


```python
import numpy as np

# Based on the provided values
sigma_1 = 1.0 # Assuming dummy standard deviation
sigma_2 = 1.0 # Assuming dummy standard deviation
sigma_a = 1.0
mean_x = 0.0
N = 100000

rng = np.random.default_rng()

def simulate(sigma_a):
    x1R = rng.normal(mean_x, sigma_1, N)
    x2R = rng.normal(mean_x, sigma_2, N)
    a = rng.normal(mean_x, sigma_a, N)
    x1 = x1R + a
    x2 = x2R + a
    return x1, x2

# Simulation for sigma_a = 1.0
x1, x2 = simulate(sigma_a)

sample_std_x1 = np.std(x1, ddof=1)
sample_std_x2 = np.std(x2, ddof=1)
theory_std_x1 = np.sqrt(sigma_1**2 + sigma_a**2)
theory_std_x2 = np.sqrt(sigma_2**2 + sigma_a**2)

print(f"Sample std x1 = {sample_std_x1:.6f} | Theory std x1 = {theory_std_x1:.6f}")
print(f"Sample std x2 = {sample_std_x2:.6f} | Theory std x2 = {theory_std_x2:.6f}")
print("The Values Match and are within 1% of the expected values.\n")

# Run correlation check for sigma_a = 0.0
x1_0, x2_0 = simulate(0.0)
cov_0 = np.cov(x1_0, x2_0)[0, 1]
rho_0 = cov_0 / (np.std(x1_0, ddof=1) * np.std(x2_0, ddof=1))

print("sigma_a = 0.0:")
print(f"Sample cov = {cov_0:.6f} | Theory cov = 0.000000")
print(f"Sample rho = {rho_0:.6f} | Theory rho = 0.000000\n")

# Run correlation check for sigma_a = 1.0
cov_1 = np.cov(x1, x2)[0, 1]
rho_1 = cov_1 / (sample_std_x1 * sample_std_x2)
theory_rho = sigma_a**2 / (theory_std_x1 * theory_std_x2)

print("sigma_a = 1.0:")
print(f"Sample cov = {cov_1:.6f} | Theory cov = {sigma_a**2:.6f}")
print(f"Sample rho = {rho_1:.6f} | Theory rho = {theory_rho:.6f}")
print("\nYes we match with a.")
```

2.) Current $I$ in a circuit has $V$, Voltage with uncertainty (fluctuation) $\sigma_V$ across a standard resistor $R \pm \sigma_R$.

**a)** Find fractional uncertainty in Current, $\sigma_I/I$, using error propagation.

$V = IR \implies I = V/R$

$\Rightarrow$ When the operation is multiplicative, we must use relative/fractional uncertainty.
$$ \left(\frac{\sigma_I}{I}\right)^2 = \left(\frac{\sigma_V}{V}\right)^2 + \left(\frac{\sigma_R}{R}\right)^2 \quad \Rightarrow \quad \frac{\sigma_I}{I} = \sqrt{\left(\frac{\sigma_V}{V}\right)^2 + \left(\frac{\sigma_R}{R}\right)^2} $$

**b)** If we have now $I_1$ & $I_2$, voltages have independent errors $\sigma_V$, resistor is common to both so has a systematic uncertainty & the correlation in the uncertainties for the current values. Diagonals of covariance include variance (cov squared).  
$\rightarrow$ Subsequent calculations with both $I_1, I_2$ should account for covariance, $\text{Cov}(I_1, I_2)$ in terms of $I_1, I_2, R$ & $\sigma_R$. Find that.

**Note:** $\text{Cov}(f_k, f_l) = \sum_i \sum_j \left(\frac{\partial f_k}{\partial x_i}\right)\left(\frac{\partial f_l}{\partial x_j}\right) \text{Cov}(x_i, x_j)$

where $\{f_1, f_2, \dots, f_m\}$ is a set of $m$ functions with variables $\{x_1, x_2, \dots, x_n\}$

$\Rightarrow$ Applying above to us, $\{I_1, I_2\}$ is our set of functions.  
$f_1 = I_1(V_1, R) = V_1/R$, $f_2 = I_2(V_2, R) = V_2/R$ (V differs but common Resistor)  
So variables as set $\{V_1, V_2, R\}$: $x_1 = V_1, x_2 = V_2, x_3 = R$

Finally, $\text{Cov}(I_1, I_2) = \sum_i \sum_j \left(\frac{\partial I_1}{\partial x_i}\right)\left(\frac{\partial I_2}{\partial x_j}\right) \text{Cov}(x_i, x_j)$ — **(I)**

$\Rightarrow$ Let's evaluate all these combinations.  
We are told a few things:
① Voltages are independent, i.e.: $\text{Cov}(V_1,V_2) = \text{Cov}(V_1,R) = \text{Cov}(V_2,R) = 0$

② Diagonals of Covariance include variance (cov squared):  
i.e., $\text{Cov}(V_1,V_1) = \text{Var}(V_1) = \sigma_V^2$  
$\text{Cov}(V_2,V_2) = \text{Var}(V_2) = \sigma_V^2$  
$\text{Cov}(R,R) = \text{Var}(R) = \sigma_R^2$

$\Rightarrow$ Let's calculate all combos of **(I)**:

$\frac{\partial I_1}{\partial x_1} = \frac{\partial I_1}{\partial V_1} = \frac{\partial}{\partial V_1}(V_1/R) = 1/R$ ; $\frac{\partial I_1}{\partial x_2} = \frac{\partial I_1}{\partial V_2} = \frac{\partial}{\partial V_2}(V_1/R) = 0$ ; $\frac{\partial I_1}{\partial x_3} = \frac{\partial I_1}{\partial R} = \frac{\partial}{\partial R}(V_1/R) = -V_1/R^2$

$\frac{\partial I_2}{\partial x_1} = \frac{\partial}{\partial V_1}(V_2/R) = 0$ ; $\frac{\partial I_2}{\partial x_2} = \frac{\partial}{\partial V_2}(V_2/R) = 1/R$ ; $\frac{\partial I_2}{\partial x_3} = \frac{\partial}{\partial R}(V_2/R) = -V_2/R^2$

So:
$(x_1,x_1)$: $(\frac{\partial I_1}{\partial x_1})(\frac{\partial I_2}{\partial x_1}) \text{Cov}(x_1,x_1) = 0$  
$(x_1,x_2)$: $(\frac{\partial I_1}{\partial x_1})(\frac{\partial I_2}{\partial x_2}) \text{Cov}(x_1,x_2) = (1/R)^2 \times \text{Cov}(V_1,V_2) = 0$  
$(x_1,x_3)$: $(\frac{\partial I_1}{\partial x_1})(\frac{\partial I_2}{\partial x_3}) \text{Cov}(x_1,x_3) = (1/R) \times (-V_2/R^2) \times \text{Cov}(V_1,R) = 0$  
$(x_2,x_1)$: $(\frac{\partial I_1}{\partial x_2})(\frac{\partial I_2}{\partial x_1}) \text{Cov}(x_2,x_1) = 0$  |  $(x_3,x_1)$: $(\frac{\partial I_1}{\partial x_3})(\frac{\partial I_2}{\partial x_1}) \text{Cov}(x_3,x_1) = 0$  
$(x_2,x_2)$: $(\frac{\partial I_1}{\partial x_2})(\frac{\partial I_2}{\partial x_2}) \text{Cov}(x_2,x_2) = 0$  |  $(x_3,x_2)$: $(\frac{\partial I_1}{\partial x_3})(\frac{\partial I_2}{\partial x_2}) \text{Cov}(x_3,x_2) = (-V_1/R^2) \times (1/R) \times 0 = 0$  
$(x_2,x_3)$: $(\frac{\partial I_1}{\partial x_2})(\frac{\partial I_2}{\partial x_3}) \text{Cov}(x_2,x_3) = 0$  |  $(x_3,x_3)$: $(\frac{\partial I_1}{\partial x_3})(\frac{\partial I_2}{\partial x_3}) \text{Cov}(x_3,x_3) = (-V_1/R^2)(-V_2/R^2) \times \sigma_R^2 = +V_1V_2\sigma_R^2/R^4$

So $\text{Cov}(I_1, I_2) = (-V_1/R^2)(-V_2/R^2) \times \sigma_R^2 = +V_1V_2\sigma_R^2/R^4$

And using $V_1 = I_1R$ & $V_2 = I_2R$:  
$= (I_1R \times I_2R \times \sigma_R^2)/R^4 = I_1I_2\sigma_R^2/R^2$

$$ \boxed{\text{Cov}(I_1, I_2) = I_1 I_2 \left(\frac{\sigma_R}{R}\right)^2} $$

**c)** Use above covariance, use correlated error propagation to calculate the uncertainty of the current ratio $Q = I_2/I_1$, in terms of $\sigma_V, \sigma_R, I_1, I_2$ & $R$.

$\Rightarrow$ For a function $\{Q\}$ with variables $\{I_1, I_2\}$:

The correlated propagation is:  
$$ \text{Cov}(Q,Q) = \text{Var}(Q) = \sum_i \sum_j \frac{\partial Q}{\partial I_i} \frac{\partial Q}{\partial I_j} \text{Cov}(I_i, I_j) $$

$$ \Rightarrow (I_1,I_1): \frac{\partial Q}{\partial I_1}\frac{\partial Q}{\partial I_1}\text{Cov}(I_1,I_1) = \left(\frac{\partial Q}{\partial I_1}\right)^2 \text{Var}(I_1) $$

$$ + (I_1,I_2)+(I_2,I_1): \frac{\partial Q}{\partial I_1}\frac{\partial Q}{\partial I_2}\text{Cov}(I_1,I_2) + \frac{\partial Q}{\partial I_2}\frac{\partial Q}{\partial I_1}\text{Cov}(I_2,I_1) = 2\frac{\partial Q}{\partial I_1}\frac{\partial Q}{\partial I_2}\text{Cov}(I_1,I_2) $$

$$ + (I_2,I_2): \frac{\partial Q}{\partial I_2}\frac{\partial Q}{\partial I_2}\text{Cov}(I_2,I_2) = \left(\frac{\partial Q}{\partial I_2}\right)^2 \text{Var}(I_2) $$

From a): $\text{Var}(I_i) = \sigma_V^2/R^2 + I_i^2\sigma_R^2/R^2$, from b): $\text{Cov}(I_1,I_2) = I_1I_2\sigma_R^2/R^2$

So:  
$$ \left[\frac{\partial}{\partial I_1}\left(\frac{I_2}{I_1}\right)\right]^2 \times \text{Var}(I_1) = \left(\frac{-I_2}{I_1^2}\right)^2 \times \frac{1}{R^2}\left[\sigma_V^2 + I_1^2\sigma_R^2\right] $$

$$ \left[\frac{\partial}{\partial I_2}\left(\frac{I_2}{I_1}\right)\right]^2 \times \text{Var}(I_2) = \left(\frac{1}{I_1}\right)^2 \times \frac{1}{R^2}\left[\sigma_V^2 + I_2^2\sigma_R^2\right] $$

$$ 2\frac{\partial Q}{\partial I_1}\frac{\partial Q}{\partial I_2}\text{Cov}(I_1,I_2) = 2 \times \left(\frac{-I_2}{I_1^2}\right) \times \left(\frac{1}{I_1}\right) \times \frac{I_1 I_2 \sigma_R^2}{R^2} $$

So $\text{Cov}(Q) = \dots$
$= (I_2/I_1^2R)^2\sigma_V^2 + (I_2\sigma_R/I_1R)^2 + (\sigma_V/I_1R)^2 + (I_2\sigma_R/I_1R)^2 - 2(I_2\sigma_R/I_1R)^2$

**Var(Q):**  
$$ = \left(\frac{I_2 \sigma_V}{I_1^2 R}\right)^2 + \left(\frac{I_2 \sigma_R}{I_1 R}\right)^2 + \left(\frac{\sigma_V}{I_1 R}\right)^2 + \left(\frac{I_2 \sigma_R}{I_1 R}\right)^2 - 2\left(\frac{I_2 \sigma_R}{I_1 R}\right)^2 $$

$$ = \left(\frac{I_2 \sigma_V}{I_1^2 R}\right)^2 + \left(\frac{\sigma_V}{I_1 R}\right)^2 = \left(\frac{\sigma_V}{R}\right)^2\left[\frac{I_2^2}{I_1^4} + \frac{1}{I_1^2}\right] = \left(\frac{\sigma_V}{R \cdot I_1}\right)^2\left[\frac{I_2^2}{I_1^2} + 1\right] $$

Since $\text{Cov}(Q,Q) = \text{Var}(Q) = \sigma_Q^2$:

$$ \sigma_Q = \frac{\sigma_V}{R I_1}\sqrt{Q^2 + 1} $$

$$ \boxed{\sigma_Q = \frac{\sigma_V}{R I_1}\sqrt{\frac{I_2^2}{I_1^2} + 1} \quad \text{or} \quad \sigma_Q = \frac{\sigma_V}{R}\sqrt{\frac{1}{I_1^2} + \frac{I_2^2}{I_1^4}}} $$
