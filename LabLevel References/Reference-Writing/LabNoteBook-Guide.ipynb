## Overview: What Good Laboratory Performance Means

Good laboratory performance means that:

- You **use your time in the lab effectively** by coming well-prepared to class, working diligently, and thinking about what you are doing
- You **spend time outside of the lab** to complete data analysis and do further reading
- You **obtain high quality data and results** from your experiment

We will assess your laboratory performance through observation, interviews, and examination of your laboratory notebook, analysis and data files.

| # | Principle | Description |

## Guidelines for Notebooks and Data Management

### The Evolution from Lab Notebook to Data Management

For centuries the primary record of any scientific activity was the bound paper notebook. Now, we rely so heavily on computers to acquire, analyze and report scientific data that a paper notebook is rarely enough to keep everything straight. The typical researcher no longer focuses exclusively on maintaining a good lab notebook, but on **good data management**. And while no system for managing scientific data is universally accepted, the basic goal of data management has not changed:

> **We need to record all of the information necessary to reproduce what we accomplish in the lab.**

Most research labs will have their own set of procedures. Here is a link to an [example](https://harmslab.uoregon.edu/wiki/harms-lab-protocols/data-management/) used by the Harms Lab at University of Oregon.


---
There are a few items to pay attention to in your lab books:

Please clearly label and date not only each in-lab session, but also all work done between lab periods (prelabs, calculations, planning, analysis). Keep entries in chronological order.

For analysis, any sample code output (e.g., a PDF export of a Python notebook) should be appended at the end. However, essential plots/results should be included in the flow of the lab book itself, with brief annotations explaining what they show.

Please clearly box or otherwise highlight main conclusions as you proceed through the lab book so they are easy to identify.

### Lab Notebook Content

Your lab notebook should be a record of the work you do on the experiment, including **preparation**, **experimental work**, and **analysis**. Think of it as a workplace journal where you record information about what you measured, how you measured it and why, and what things you did to improve your measurement.

#### Required Details

Be sure to include details such as:

- [ ] Date, names, what aspect of the lab you are working on
- [ ] Equations that you use
- [ ] Sketches and photos of apparatus
- [ ] Lists of equipment
- [ ] Location of files
- [ ] Rough plots and final plots
- [ ] Things that worked and things that did not

#### Important Notes

- It should be an **up to date** account of what you are working on
- You do **not** need to go back and improve things or add things—just start a new section with a date and topic
- We do not expect extended essays in your lab notebook
- We expect your notes to be **comprehensive enough** that:
  - You could use them months later when writing up your project
  - Another student could use them to figure out what you had done so that they could replicate your results

#### Assessment Focus: Reproducibility

In our assessment, we will focus on **reproducibility**:

- How well could your experiment be reproduced by another student based on your lab notebook?
- Would they be able to find all the data?
- Would they understand how you set up your apparatus or analyzed your data?

#### Paper vs. Electronic

Paper still has many advantages over computerized documentation: it is fast, cheap, flexible, and durable. Unless you have already established a practice of taking notes electronically, use a **bound laboratory notebook** where the pages are numbered. If you use an electronic method, make sure that you can export the contents to a PDF for assessment (and archiving, if you are using it in a research context).

#### Scanner

If you use a paper notebook you will need to use a digital scanner to convert it to an electronic format. The library has several printer/copier/scanner machines available for student use, and there are a variety of scanner apps available for tablets and smartphones.

### Managing Data and Analysis

Each group should create its own shared file space on Canvas. Store all of your data and analysis files here so that you, other group members, and your instructors have access to it. (Even if doing an experiment on your own, you should follow these guidelines.)

#### File Formats

Consider the file formats that you will use for your data and analysis. For your formal report in LaTeX, all imported figures are easiest to handle if converted to [PDF](https://en.wikipedia.org/wiki/PDF) files. You should figure out how to do this on your computer (e.g., by printing or exporting to PDF).

For long-term storage, **non-proprietary formats** may be more stable:

| Data Type | Recommended Format |
|-----------|-------------------|
| Simple tabular data | [CSV](https://en.wikipedia.org/wiki/Comma-separated_values) (comma separated value) |
| Images | [PNG](https://en.wikipedia.org/wiki/Portable_Network_Graphics) or [JPEG](https://en.wikipedia.org/wiki/JPEG) |
| Vector graphics (with text) | [SVG](https://en.wikipedia.org/wiki/Scalable_Vector_Graphics) |

#### Analysis Scripts

- Organize your data analysis in a computer script and upload it to your Canvas group file space
- Update it as you develop it
- Be sure to include any information that would be necessary to reproduce plots that you include in your notebook
- This may be a script, or just a copy of the commands that you used to produce it

### Organizing Your File Space

Please organize your directory and name your files according to the following template.

At the top directory for each experiment you should include a **README file** that briefly summarizes the other files in the directory.

```
<ExperimentNameA>/              # top directory for the experiment
├── README.txt                  # Readme file
├── <Analysis>/                 # Analysis directory
│   ├── finalAnalysisScript.py
│   ├── myfitfun.py
│   └── preliminaryAnalysisScript.py
├── <Data>/                     # Data directory
│   ├── <2018-01-16>/
│   │   ├── 2018-01-16-descriptive-title.csv
│   │   └── 2018-01-16-new-descriptive-title.csv
│   └── <2018-01-23>/
│       ├── 2018-01-23-descriptive-title.csv
│       └── 2018-01-23-new-descriptive-title.csv
└── <Notes>/                    # Notes and other documents
    ├── apparatus-pic.png
    └── schematic.svg
```

### README Files

The purpose of a readme file is to record **metadata** about a set of files and folders.

#### Why README Files Matter

Imagine doing an experiment, and recording data, notes, and analysis files into some top level folder:

```
MyFantasticLab/
├── Analysis/
├── Data/
└── Notes/
```

What if you were then to send this information to a collaborator, as a zipped file, with name `<LabName>_<YourName>.zip`. That collaborator, being unorganized, decides to save the file as `downloaded_file.zip` and then forgets about it for a few weeks. When they come back and try to understand the contents of that `downloaded_file.zip`, they will have almost certainly forgotten the context of the data. Worse, they might even forget who sent the data, and think that they created this and therefore own the rights to the data!

One way to help prevent this kind of confusion is to **include a README file in every project top-level directory**.

#### README Template

The exact format of the file is up to you, but the goal is always the same—to convey as much information as possible about the data you are storing in the files.

```
<Project Name>
<Project Description>
<Project Purpose>
<Date when project was started/completed>

Names of collaborators:
- <Name 1>
- <Name 2>

Requirements (if there are specific things needed to re-create the analysis):
- Python 3.6
- Jupyter lab (syzygy)
- SciPy <version>
- <Name of obscure analysis package and where to get it from>

Important files:
- <filename 1>: <Description of why this file is particularly important>
- speed_of_sound_2019-03-19.csv: Final dataset for the speed of sound of air. 
  The lab temperature was 20.3 C during data collection.
```

### Thinking Long Term

In a research context, it is common to need access to data and notes that were acquired years earlier. You probably will not need your records for this course after the term is over, but we will assess you as if you did.

**Best Practices:**

- Write clearly and legibly
- Organize your electronic files well
- Use electronic file formats that you know you will be able to read ten years from now
- Make sure to capture all of the information necessary to reproduce your experiment
- **Back up anything that is in electronic format!** Electronic files are notoriously ephemeral

> Just ask [this Melbourne man](https://www.abc.net.au/news/2018-01-17/bitcoin-stored-on-hard-drive-lost-when-sent-to-a-tip/9335490) who had a digital wallet with the keys to thousands of Bitcoins stored on a thumb drive that became corrupted. Had he kept better backups, he could have sold these at the end of 2017 for more than $10M.

### Snapshot Submissions (Week 1 & Week 2)

Following completion of a set of lab sessions (generally after Week 1 and Week 2), the lab notebook will be evaluated as a "snapshot" of the work completed during the lab.

#### What to Submit

Submission should include:
- All pages added to your lab book since the last submission
- Any figures, plots, tables you wish to add
- All collated into a **single PDF**

#### Required: Submission Summary

For each of the Week 1 and Week 2 submissions, please add at the end a **short summary list of items done during that week with page number references** pointing to the relevant part of the notebook—essentially a glorified table of contents to guide the grader.

#### Snapshot Evaluation Criteria

This evaluation will be based on four major components:

---

| Level | Description |
|-------|-------------|
| **Proficient** | Pre-lab questions are complete and correct; initial calculations or simulations show that the student has thought about what to expect; the goals for each lab period or analysis session are specific and stated clearly at the beginning of each lab or analysis session; the description of proposed procedure is clearly stated and is informative |
| **Satisfactory** | Pre-lab questions have been attempted; preparation is adequate; specific goals are included, basic procedure is proposed |
| **Insufficient** | Pre-lab work is minimal or non-existent |

##### 2. Progress

| Level | Description |

### Final Lab Book Submission

Following the end of an experiment there will be additional time to do a full and proper analysis of the experiment.

#### What to Submit

This should include:
- All relevant figures, images
- A discussion of results
- A proper conclusion to the lab
- Every page of your lab book relevant to the experiment
- Any figures, plots, tables
- All collated into a **single PDF**

#### Required: Executive Summary

You need to include at the end of the submission an **executive summary** of the main figures, results and conclusions with page references to the appropriate sections of your lab book where the full details may be found.

> **Important:** Do not reproduce figures and tables in the summary; just provide page references.

To emphasize:
- Don't change what you are doing from day to day
- Do not turn the executive summary into a big project or make a write-up
- You just have to gather together the main items and boxed conclusions from your lab book pages together with some summary discussion for context
- Page references will allow the grader to go back and assess the full details in your lab book and allow you to keep the summary brief

## Quick Reference: Notebook Checklist

### Every Session Must Include

- [ ] Date and time
- [ ] Partner name(s) and roles
- [ ] Session goals (specific, measurable)
- [ ] Independent, dependent, and control variables defined
- [ ] Equipment list with relevant settings
- [ ] Procedure with timestamps
- [ ] Data files saved with descriptive names
- [ ] File locations recorded
- [ ] Observations and troubleshooting notes
- [ ] Figures/plots of data taken during session
- [ ] End-of-session reflection on goals

### Every Data Table Must Include

- [ ] Clear labels and units
- [ ] Uncertainties (as 1-sigma values)
- [ ] Independent variable in first column
- [ ] Dependent variable(s) in subsequent columns

### Every Figure Must Include

- [ ] Labeled axes with units
- [ ] Error bars
- [ ] Legend (if multiple datasets)
- [ ] Fit line (if applicable)
- [ ] Residuals (if fitting)
- [ ] Caption or description

### Uncertainty Analysis Must Include

- [ ] Random uncertainties identified
- [ ] Systematic uncertainties identified
- [ ] Dominant uncertainty source identified
- [ ] Propagation to final results
- [ ] Impact on conclusions discussed

### Final Results Must Include

- [ ] Value ± uncertainty (1-sigma)
- [ ] Comparison to theory/known values
- [ ] Percent discrepancy calculated
- [ ] Statement of agreement within uncertainty
